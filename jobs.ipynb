### **About Dataset**

  `work_year` : The year in which the data was recorded. \
  `job_title`: The specific title of the job role, like 'Data Scientist', 'Data Engineer', or 'Data Analyst'. This column is crucial for understanding the salary distribution across various specialized roles within the data field.\
`job_category`: A classification of the job role into broader categories for easier analysis. This might include areas like 'Data Analysis', 'Machine Learning', 'Data Engineering', etc.\
`salary_currency`: The currency in which the salary is paid, such as USD, EUR, etc. This is important for currency conversion and understanding the actual value of the salary in a global context.\
`salary`: The annual gross salary of the role in the local currency. This raw salary figure is key for direct regional salary comparisons.\
`salary_in_usd`: The annual gross salary converted to United States Dollars (USD). This uniform currency conversion aids in global salary comparisons and analyses.\
`employee_residence`: The country of residence of the employee. This data point can be used to explore geographical salary differences and cost-of-living variations.\
`experience_level`: Classifies the professional experience level of the employee. Common categories might include 'Entry-level, 'Mid-level', 'Senior', and 'Executive', providing insight into how experience influences salary in data-related roles.\
`Employment_type`: Specifies the type of employment, such as 'Full-time', 'Part-time', 'Contract', etc. This helps in analyzing how different employment arrangements affect salary structures.\
`work_setting`: The work setting or environment, like 'Remote', 'In-person', or 'Hybrid'. This column reflects the impact of work settings on salary levels in the data industry.\
`company_location`: The country where the company is located. It helps in analyzing how the location of the company affects salary structures.\
`company_size`: The size of the employer company, often categorized into small (S), medium (M), and large (L) sizes. This allows for analysis of how company size influences salary.

### **Apply groupby function for more understanding the data**

Create hyphothesis

```python
df.columns
```

```python
#Salary distribution across job categories
hypothesis_1 = df.groupby('job_category')['salary'].mean().sort_values(ascending=False)

# Plotting the hypothesis
plt.figure(figsize=(12, 6))
sns.barplot(y=hypothesis_1.index, x=hypothesis_1.values,color='#c3bd30')
plt.title('Average Salary Across Job Categories')
plt.xlabel('Job Category')
plt.ylabel('Average Salary')
plt.show()
```

`This hypothesis explores whether there are significant differences in average salaries across different job categories.`

```python
# Hypothesis 2: Relationship between experience level and salary
hypothesis_2 = df.groupby('experience_level')['salary'].mean().sort_values(ascending=False)

# Plotting the hypothesis
plt.figure(figsize=(10, 6))
sns.barplot(x=hypothesis_2.index, y=hypothesis_2.values,color='#a4c330')
plt.title('Median Salary Across Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Median Salary')
plt.show()
```

```python
#Company Size and Employment Type Distribution
hypothesis_3 = df.groupby(['company_size', 'employment_type']).size().unstack()

# Plotting the hypothesis
hypothesis_3.plot(kind='bar', colormap='coolwarm')
plt.title('Distribution of Employment Types Within Company Sizes')
plt.xlabel('Company Size')
plt.ylabel('Count')
plt.show()
```

```python
df['company_size'].value_counts()
```

```python
 #Average Salary Across Different Work Settings
hypothesis_4 = df.groupby('work_setting')['salary_in_usd'].mean().sort_values(ascending=False)

# Plotting the hypothesis
plt.figure(figsize=(12, 6))
sns.barplot(x=hypothesis_4.index, y=hypothesis_4.values, color='#181e31')
plt.title('Average Salary Across Work Settings')
plt.xlabel('Work Setting')
plt.ylabel('Average Salary (USD)')
plt.show()
```

`This hypothesis explores whether there are significant differences in average salaries across different work setting`

```python
#Salary Distribution by Employement_type
hypothesis_5 = df.groupby('employment_type')['salary'].median().sort_values(ascending=False)

# Plotting the hypothesis
plt.figure(figsize=(14, 6))
sns.barplot(x=hypothesis_5.index, y=hypothesis_5.values)
plt.title('Median Salary Across Employement Type')
plt.xlabel('Job Employement type')
plt.ylabel('Median Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
#Relationship Between Job Category and Work Setting
hypothesis_6 = df.groupby(['job_category', 'work_setting']).size().unstack()

# Plotting the hypothesis
hypothesis_6.plot(kind='bar', stacked=True, colormap='Pastel1', figsize=(14, 6))
plt.title('Relationship Between Job Category and Work Setting')
plt.xlabel('Job Category')
plt.ylabel('Count')
plt.show()
```

### **Corrrelation**

```python
# Calculate the correlation matrix
correlation_matrix = df[['work_year','salary','salary_in_usd']].corr()
# Plot a heatmap of the correlation matrix
#plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True,  cmap='coolwarm',linewidths=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plot for the correlation between 'work_year' and 'salary_in_usd'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='work_year', y='salary_in_usd', data=df)
plt.title('Scatter Plot: work_year vs Salary in USD')
plt.xlabel('work_year')
plt.ylabel('Salary in USD')
plt.show()
```
